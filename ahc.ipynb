```python
# import Library 
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
import pandas as pd
```

```python
#import dataset kedalam variabel x, dataset ini yang akan kita cluster
x = pd.read_csv('datanilai.csv')
print(x)
```

```python
#lakukan hierachical klastering
z = linkage(x)
```

```python
print(z)
```

```python
#plot dendogram
plt.figure(figsize=(25, 10))
plt.title('Dendogram Hierarchical Klastering')
plt.xlabel('SAT')
plt.ylabel('GPA')
dendrogram(z)
plt.show()
```
